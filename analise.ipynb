```python
import pandas as pd
import re

pd.set_option('display.max_columns', None)
jogadores = pd.read_excel('C://Users//eduar//Desktop//DEV//Football//MLS//jogadores_mls.xlsx', index_col= None)
database = pd.read_csv('C://Users//eduar//Desktop//DEV//Football//MLS//database.csv', delimiter=',')
```

## Inicial

```python
def remove_lowercase(value):
    if isinstance(value, str):
        return re.sub(r'[a-z]', '', value)
    return value

# Aplicar a função à coluna 'Nação'
database['Nação'] = database['Nação'].apply(remove_lowercase)

# Ordenando por data
database = database.sort_values(by='Data',ascending= True)

database = pd.merge(database, jogadores[['Jogador','Equipe']], on='Jogador', how='left')

colunas = ['Equipe'] + [col for col in database.columns if col != 'Equipe']
database = database[colunas]
```

```python
database.groupby(by='Jogador').agg(Aparições = ('Jogador','count')).sort_values(by='Aparições', ascending= False)
```

## Agrupado

```python
# Agrupar os dados por jogador e calcular a soma dos chutes ao gol e a contagem de aparições
db_agrupado = database.groupby(['Jogador','Equipe']).agg(
    Total_Gols=('Gols','sum'),
    Total_Chutes=('TC', 'sum'),   # Soma dos chutes ao gol
    Aparicoes=('Jogador', 'count'), # Contagem de aparições
    Total_Chutes_Gol=('CaG', 'sum'), # Contagem de aparições
).sort_values(by='Total_Chutes', ascending=False)  # Ordenar antes do reset_index
```

```python
db_agrupado['Média_chute']=round(db_agrupado['Total_Chutes'] / db_agrupado['Aparicoes'],2)
db_agrupado['Média_gol']=round(db_agrupado['Total_Gols'] / db_agrupado['Aparicoes'],2)
```
